```python
def print_recursive(elem: Element, parents=None):
    parents = parents.copy() if parents else []
    elem = elem

    parents.append(elem.tag.replace('{http://www.portalfiscal.inf.br/nfe}', ''))

    print(len(elem), ':'.join(parents), elem.attrib, elem.text.replace('\n', '') if elem.text else None)

    if len(elem):
        for child in elem:
            print_recursive(child, parents)

print_recursive(root[0][0])
```

```python
# Import BeautifulSoup
from bs4 import BeautifulSoup as bs
import os
import csv

content = []
fileName = 'drive/MyDrive/XML/NFe35170271322150003428550000000386591831242542_procNFe.xml'
outFileName1 = 'drive/MyDrive/XML/temp1.csv'
delimiter = "|"
_urlNFE ='http://www.portalfiscal.inf.br/nfe'

# Read the XML file
with open(fileName, "r") as file:
    # Read each line in the file, readlines() returns a list of lines
    content = file.readlines()
    # Combine the lines in the list into a string
    content = "".join(content)
    bs_content = bs(content, "lxml")

def CreateHeader( content:bs, section:str , delimiter:str='|' ):
  header = ''
  for elem in content.find(section):
    header +=  elem.name + delimiter

  header = header[:-1]
  return (header)

def CreateRow( content:bs, section:str , delimiter:str='|' ):
  row = ''
  for elem in content.find(section):
    row +=  elem.text + delimiter

  row = row[:-1]
  return(row)

def AddNode( path:str) -> str:
    return '{' + _urlNFE + '}' + path


def GetElement( doc , path:str, elemName:str ):
  nodefind = doc.find( path )
  result = nodefind.attrib.get( elemName ) 
  return (result)

def GetSubElement( doc , path:str, elemName:str ):
  path += '/{' + _urlNFE + '}'+ elemName
  nodefind = doc.find( path )
  result = nodefind.text
  return (result)


h= CreateHeader(bs_content,'ide')
r= CreateRow(bs_content , 'ide')

print( h , '\n' , r)
```
